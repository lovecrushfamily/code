#### Pie chart

```python
labels = [
    'Children 0-18', 'Adults 19-25', 'Adults 26-34', 
    'Adults 35-54', 'Adults 55-64', 'Seniors 65+'
]
 
data = [
    75307800, 27799100, 39817700, 
    81478600, 42061700, 52784400
]


# Color Palette
Spring_Pastels = ["#fd7f6f", "#7eb0d5", "#b2e061", "#bd7ebe", "#ffb55a", "#ffee65", "#beb9db", "#fdcce5", "#8bd3c7"]

# configuration

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} )".format(pct, absolute)

plt.figure(figsize=(8,8))
tuple = plt.pie(
    x=data, 
    labels=labels,

    autopct=lambda pct: func(pct, data),
    # autopct='%1.2f%%', # show percentage with two decimal points
    textprops={'fontsize':11}, # increase the size of all text elements 
    colors=Spring_Pastels[:len(data)],

    startangle=90, 
    counterclock=False,
    
    labeldistance=0.8,
    pctdistance=0.45,

    explode = list(map(lambda x : x if (x < 0.15) else 0.01, np.random.random(len(data))))
)

for text in tuple[1]: #texts
    text.set_fontweight('regular')
    text.set_horizontalalignment('center')

for auto_text in tuple[2]: # Autotexts
    auto_text.set_horizontalalignment('center')
    auto_text.set_fontstyle('italic')

plt.legend(tuple[0], labels,
        title="Type",
        loc="center left",
        bbox_to_anchor=(1, 0, 0, 1.5))

plt.title(
    label="Name of this chart", # Label
    fontdict={"fontsize":17},
    loc='center'
)




plt.show()
```

#### Subplot pie Chart

```python
 
# Creating dataset
labels = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
 

#  Configuration 
Spring_Pastels = ["#fd7f6f", "#7eb0d5", "#b2e061", "#bd7ebe", "#ffb55a", "#ffee65", "#beb9db", "#fdcce5", "#8bd3c7"]

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(x=data,
                                  autopct=lambda pct: func(pct, data),
                                  explode=list(map(lambda x : x if (x < 0.24) else 0.08, np.random.random(len(data))))   ,
                                  labels=labels,
                                  shadow=True,
                                  colors=(Spring_Pastels)[:len(data)],
                                  startangle=90,
                                  wedgeprops= {'linewidth': 1, 'edgecolor': "slategray"}, # border
                                  textprops=dict(color="slateblue"))
ax.legend(wedges, labels,
          title="Cars",
          loc="center left",
          bbox_to_anchor=(1.1, 0, 0, 1.5))
# ax.legend()
 
plt.setp(autotexts, size=10, weight="bold")
ax.set_title("Customizing pie chart")
 
plt.show()
```

#### Bar Chart (Single column)

```python
# custom

# creating the dataset
data = {'C':20, 'C++':15, 'Java':30, 
        'Python':70}
groups = list(data.keys())
values = np.array(list( data.values()))

width = 0.8
height = 6
  
plt.figure(figsize = (7,5), dpi=100)
 
plt.bar(x=groups, 
        height= values, 
        # color ='cornflowerblue', 
        color=sns.color_palette('Set2', len(data)),       
        width = width,
        tick_label = groups,
        edgecolor='royalblue')

for group, value in zip(groups, values):
    plt.text(group, value + 1, f'{value}', ha='center',color='cornflowerblue')
      
# plt.xticks(rotation=60)

plt.tight_layout()
plt.xlabel("x-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15}, labelpad=10)
plt.ylabel("y-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15},labelpad=10)
plt.title("Your chart name here",fontdict={'color':'royalblue','family':'serif','size':17},loc='center',pad=20)
plt.legend()
plt.show()
```

#### Pattern bar chart

```python
# creating the dataset

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-light.mplstyle')

data = {'C':20, 'C++':15, 'Java':30, 
        'Python':70}
groups = list(data.keys())
values = np.array(list( data.values()))

width = 0.8
height = 6
  
plt.figure(figsize = (7,5), dpi=100)
 
bar_ = plt.bar(x=groups, 
        height= values, 
        color ='cornflowerblue',        
        width = width,
        tick_label = groups,
        edgecolor='royalblue')

# set pattern for column 
patterns = ['/', 'O', '*', '|']
for bar in bar_:
    bar.set_hatch(patterns.pop())

for group, value in zip(groups, values):
    plt.text(group, value + 1, f'{value}', ha='center')

plt.xticks(rotation=60)

plt.xlabel("x-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15})
plt.ylabel("y-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15})
plt.title("Your chart name here",fontdict={'color':'royalblue','family':'serif','size':17},loc='center')
plt.show()

```

#### Bar Chart (multi-column)

```python

barWidth = 0.1


 
# set height of bar 
IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 
labels_x_axis = ['2015', '2016', '2017', '2018', '2019']
 
plt.figure(figsize=(8,5), dpi=100)
# Set position of bar on X axis 
br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
 
# Make the plot
plt.bar(br1, IT, color ='slateblue', width = barWidth,  edgecolor ='grey',align='edge', label ='IT') 
plt.bar(br2, ECE, color ='dodgerblue', width = barWidth, edgecolor ='grey',align='edge', label ='ECE') 
plt.bar(br3, CSE, color ='skyblue', width = barWidth, edgecolor ='grey',align='edge', label ='CSE') 

# Adding Xticks 
plt.xticks([r + barWidth for r in range(len(IT))], labels_x_axis)

plt.grid(True, linewidth=1, linestyle='--') #only for single column

plt.xlabel("x-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15}, labelpad=10)
plt.ylabel("y-axis name",fontdict={'color':'deepskyblue','family':'serif','size':15},labelpad=10)
plt.title("Your chart name here",fontdict={'color':'royalblue','family':'serif','size':17},loc='center',pad=20)
plt.legend()

plt.show() 
```

#### Stacked Bar plot



```python
N = 5
 
boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)   
width = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,bottom = boys, yerr = girlStd)
 
plt.title('The chart title')
plt.ylabel('Contribution')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))
 
plt.show()
```

#### Smooth 2DLine plot

```python

import numpy as np

T = np.array([6, 7, 8, 9, 10, 11, 12])
power = np.array([1.53E+03, 5.92E+02, 2.04E+02, 7.24E+01, 2.72E+01, 1.10E+01, 4.70E+00])

plt.plot(T,power)
plt.show()

from scipy.interpolate import make_interp_spline, BSpline

# 300 represents number of points to make between T.min and T.max
xnew = np.linspace(T.min(), T.max(), 300) 

spl = make_interp_spline(T, power, k=3)  # type: BSpline
power_smooth = spl(xnew)

plt.plot(xnew, power_smooth)
plt.show()
```
